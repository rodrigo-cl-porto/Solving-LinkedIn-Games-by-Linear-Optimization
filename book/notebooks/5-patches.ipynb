# Problem Modeling

## Ranges

$I = {1, \cdots, n}$
: The row range, where $n$ is the total amount of rows on game board.

$J = {1, \cdots, m}$
: The column range, where $m$ is the total amount of columns on game board.

## Sets

$H = I \times J = \{(i,j) \mid \forall i \in I, \forall j \in J\}$
: Set of all board squares, that is just the Cartesian product between sets $I$ and $J$.

$T \subset H$
: Set of board cells with tips about a rectangle $k$.

$K = {1, \cdots, p}$
: Set of all rectangles to be drawn on board game.

$V \subseteq K$
: Set of vertical rectangles, i. e., rectangles whose height is greater than its width.

$O \subseteq K$
: Set of horizontal rectangles, i. e., rectangles whose width is greater than its height.

$S \subseteq K$
: Set of square figures, i. e., rectangles whose width is equal to its height.

## Decision Variables

$a_{ijk} \in \{0,1\}, \forall (i,j,k) \in I \times J \times K$
: Decision to cover the cell $(i,j)$ with the rectangle $k$ or not.

$b_{ik} \in \{0,1\}, \forall (i,k) \in I \times K$
: $b_{ik}=1$ if the row $i$ is part of rectangle $r$; $b_{ik}=1$, otherwise.

$c_{ik} \in \{0,1\}, \forall (j,k) \in J \times K$
: $b_{jk}=1$ if the column $j$ is part of rectangle $r$; $c_{jk}=1$, otherwise.

$x_k \in I, \forall k \in K$
: The column index of the upper leftmost cell of the rectangle $k$.

$y_k \in J, \forall k \in K$
: The row index of the upper leftmost cell of the rectangle $k$.

$w_k \in I, \forall k \in K$
: Width of the rectangle $k$.

$h_k \in J, \forall k \in K$
: Height of the rectangle $k$.

## Parameters

$a_k \in mathbb{N}$
: Predefined area for the rectangle $k$.

## Objective function

$\text{Max} \ C$

## Constraints

$\sum_{k \in K}{a_{ijk} = 1, \forall (i,j) \in H}$
: Each square board must be covered by only one rectangle $k$.

$x_k - j \le M(1-a_{ijk})$
: if a cell $(i,j)$ is covered by rectangle $k$, then its $j$ index must be greater or equal than the column index of $k$'s first square.

$j - (x_k + w_k - 1) \le M(1-a_{ijk})$
: if a cell $(i,j)$ is covered by rectangle $k$, then its $j$ index must be less or equal than the column index of $k$'s last square.

$y_k - i \le M(1-a_{ijk})$
: if a cell $(i,j)$ is covered by rectangle $k$, then its $i$ index must be greater or equal than the row index of $k$'s first square.

$i - (y_k + h_k - 1) \le M(1-a_{ijk})$
: if a cell $(i,j)$ is covered by rectangle $k$, then its $j$ index must be less or equal than the column index of $k$'s last square.
